**Task 1**: Plot the distribution of combined fuel mileage (column 'comb', in miles per gallon) by manufacturer (column 'make'), for all manufacturers with at least eighty cars in the dataset. Consider which manufacturer order will convey the most information when constructing your final plot. **Hint**: Completing this exercise will take multiple steps! Add additional code cells as needed in order to achieve the goal.

```python
# Get top 18 manufactured cars
top_cars = fuel_econ['make'].value_counts().index[:18]
fuel_econ_top18 = fuel_econ[fuel_econ['make'].isin(top_cars)]

# Sort by each group comb mean
mean_comb_order = fuel_econ_top18.groupby('make')['comb'].mean().sort_values(ascending=False).index

# Plot facet grid with histograms
g = sb.FacetGrid(data=fuel_econ_top18, col="make", col_wrap=6, col_order=mean_comb_order)
g = g.map(plt.hist, "comb")
g.set_titles("{col_name}")
plt.show()
```

```python
# Get top 18 manufactured cars
top_cars = fuel_econ['make'].value_counts().index[:18]
fuel_econ_top18 = fuel_econ[fuel_econ['make'].isin(top_cars)]

# Sort by each group comb mean
mean_comb_order = fuel_econ_top18.groupby('make')['comb'].mean().sort_values(ascending=False).index

# Plot facet grid with histograms
adjust=2
bins=np.arange(0, fuel_econ_top18['comb'].max()+adjust, adjust)
g = sb.FacetGrid(data=fuel_econ_top18, col="make", col_wrap=6, col_order=mean_comb_order)
g = g.map(sb.distplot, "comb", kde=False, bins=bins)
g.set_titles("{col_name}")
plt.show()
```

**Task 2**: Continuing on from the previous task, plot the mean fuel efficiency for each manufacturer with at least 80 cars in the dataset.

```python
# Get dataset with at least 80 cars for each manufacturer
cars_count = fuel_econ['make'].value_counts()
cars_sub = cars_count[cars_count >= 80].index
data_sub = fuel_econ[fuel_econ['make'].isin(cars_sub)]

# Plot barplot
sb.barplot(y="make", x="comb", data=data_sub, color='#3274A1')
plt.show()
```

```python
# Add sorting by comb mean
# Set confidence interval = standard deviation

# Get dataset with at least 80 cars for each manufacturer
cars_count = fuel_econ['make'].value_counts()
cars_sub = cars_count[cars_count >= 80].index
data_sub = fuel_econ[fuel_econ['make'].isin(cars_sub)]

# Plot barplot
display_order = data_sub.groupby(['make']).mean().sort_values('comb', ascending=False).index
sb.barplot(y="make", x="comb", data=data_sub, color='#3274A1', order=display_order, ci='sd')
plt.show()
```
